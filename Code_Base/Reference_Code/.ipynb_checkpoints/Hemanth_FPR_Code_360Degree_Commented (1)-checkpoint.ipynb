## Load in tensorflow and preprocessing

```python
# Image Preprocesing (Augmentation)
# Image data cannot be accessed directly , so it need to be encoded to numerical format
# augmentation is the process through whcih the image can be converted to image tensor array which contain the image pixel values
# to prepare the augmentation model (keras.io and tensorflow), several prametesr can eb used and some of thsoe have been applied by emphasizing the image rotation values
# image rotation has been set to three different values(as three notebooks provided)
# rescale operatiuon has bene described in FPR in the image augmentation section with example 

train_datagen = keras.preprocessing.image.ImageDataGenerator(rescale=1. / 255,
                                   rotation_range=360,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True,
                                   fill_mode='nearest')

test_datagen = keras.preprocessing.image.ImageDataGenerator(rescale=1. / 255)

BSZ = 16
HgtImg = 128
WdhImg = 128
```

```python
# preparingb augmented training database
train_ds = keras.preprocessing.image_dataset_from_directory(
  drtr,
  image_size=(HgtImg, WdhImg),
  batch_size=BSZ)
train_generator = train_datagen.flow_from_directory(
        drtr,
        classes=CovClass,
        target_size=(HgtImg, WdhImg),
        batch_size=32,
        class_mode='categorical')
```

```python
# preparingb augmented test database
test_ds = keras.preprocessing.image_dataset_from_directory(
    drts,
    labels="inferred",
    image_size =(HgtImg, WdhImg),
    batch_size = BSZ)
test_generator = test_datagen.flow_from_directory(
        drts,
        classes=CovClass,
        target_size=(HgtImg, WdhImg),
        batch_size=32,
        class_mode='categorical')
```

```python
def CreateData(MDLHST, MName,itr):     # function to prepare the dataset for training and validation performances (accuracy and losses)
    CurrHist=MDLHST.history
    DFCovDetect=pandas.DataFrame({
        "Epoch":[i+1 for i in range(itr)],
        "Loss(Training)":CurrHist['loss'],
        "Loss (Test)":CurrHist['val_loss'],
        "Accuracy(Training)":CurrHist['accuracy'],
        "Accuracy (Test)":CurrHist['val_accuracy']
    })
    DFCovDetect.to_csv("DFCovDetect_{}.csv".format(MName))    # store the dataframe for future use
    print(DFCovDetect.head())
    return DFCovDetect
```

```python
wgt="imagenet"
inctop=False

# accessing the Transfer Learning models (keras.io)
CovModels=[keras.applications.DenseNet201(input_shape=[HgtImg, WdhImg] + [3],weights=wgt, include_top=inctop),
          keras.applications.EfficientNetB2(input_shape=[HgtImg, WdhImg] + [3],weights=wgt, include_top=inctop),
          keras.applications.VGG16(input_shape=[HgtImg, WdhImg] + [3],weights=wgt, include_top=inctop),
          keras.applications.Xception(input_shape=[HgtImg, WdhImg] + [3],weights=wgt, include_top=inctop)
          ]
CovModelNames=["Densenet","EfficientNetV2","VGG16","Xcepotion"]    # names of the models


CovACCtr,CovLosstr,CovACCts,CovLossts=[],[],[],[]     # lists to store the training and testing accuracies and losses (Train accuracy, Train loss, Test accuracy, Test loss)
TrItr=50    # epoch value for training (fit)
ColScheme=["Wistia_r","bwr","jet","mako"]
CovModelToStore=[]

lossfunc='categorical_crossentropy'    # loss function-> this is chosen as the labesl of the data are categorical and more than two-> covid, pneumonia and normal
                                       # -> will show training and validation losses
metcontrol='accuracy'    # performance control metrics -> will show training and validation accuracies
optfucn='adam'        # optimizer is adam (keras.io) for model compilation-> Adaptive Moment Estimation (Adam)-> Adam adjusts the learning rate at each iteration to improve convergence
ipact='relu'          # actuivation fucntion for transafer learning model (keras.io) -> ReLu, or Rectified Linear Unit, is a simple and effective activation function that sets all negative values to zero
opact='softmax'       # activation fucntion for model at the output layer, it will compute the porbabilities of each classes

def CovModelFunction(CMDL):    # transfer learning model design
  modCov = CMDL
  for mlyr in modCov.layers:
      mlyr.trainable=False
  CovMdlLayer=keras.layers.Flatten()(modCov.output)      # concatenating flatten layer for transform n-D data to 1-D data for faster processing
  covpred=keras.layers.Dense(3,activation=opact)(CovMdlLayer)     # assigning output layer with 3 neurones as there are 3 classes
  MovelCov_Pres = keras.models.Model(inputs=modCov.input, outputs=covpred)   # creating complete model by concatenating input and output layers
  # model compilation
  MovelCov_Pres.compile(
    loss=lossfunc,
    optimizer=optfucn,
    metrics=[metcontrol]
  )
  MovelCov_Pres.summary()
  return MovelCov_Pres

for i in range(len(CovModels)):
    print("=================================================================================================")
    print("                               Executing {} Model".format(CovModelNames[i]))
    print("=================================================================================================")
    MovelCov_Pres = CovModelFunction(CovModels[i])    # calling model creating, function for each of the transfer learning algorithms 
                                                      # (using the loop where in eachbiteartion, one model will be created)
    # model training
    TrainCovModel = MovelCov_Pres.fit(train_generator,
                   epochs=TrItr,
                   verbose=1,
                   validation_data=test_generator)
    ModelHist=CreateData(TrainCovModel,CovModelNames[i],TrItr)     # create the data for training and validation performanecs
    
    CovModelToStore.append(MovelCov_Pres)

    print("-.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.-.--")
    print("                    Train Performance for {} Model".format(CovModelNames[i]))
    print("-.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.-.--")
    LossFinCovtr, AccFinCovtr = MovelCov_Pres.evaluate(train_generator)      # evaluate the model using training data
    print('Final Loss: {}, Final Accuracy: {}%'.format(LossFinCovtr, round(AccFinCovtr,3)*100))
    CovACCtr.append(round(AccFinCovtr,3)*100)     # store the training accuracy
    CovLosstr.append(round(LossFinCovtr,3))     # store the training loss

    print("-.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.-.--")
    print("                    Test Performance for {} Model".format(CovModelNames[i]))
    print("-.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.-.--")
    LossFinCov, AccFinCov = MovelCov_Pres.evaluate(test_generator)         # evaluate the model using test data
    print('Final Loss: {}, Final Accuracy: {}%'.format(LossFinCov, round(AccFinCov,3)*100))
    CovACCts.append(round(AccFinCov,3)*100)         # store the test accuracy
    CovLossts.append(round(LossFinCov,3))     # store the test loss
    tlcol="#DC143C"
    sid,eid=1,3
    print("-.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.-.--")
    print("                    Performance Graph for {} Model".format(CovModelNames[i]))
    print("-.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.--.-.-.--")
    ModelHist.iloc[:,eid:].plot(kind="line",figsize=(12,7),color=["#342D7E","#008000"])
    pyplot.title("Model Accuracies for {}".format(CovModelNames[i]),fontsize=20,color=tlcol)
    pyplot.xlabel("Epoch",fontsize=20,color=tlcol)
    pyplot.ylabel("Accuracies",fontsize=20,color=tlcol)
    pyplot.grid()
    pyplot.show()

    ModelHist.iloc[:,sid:eid].plot(kind="line",figsize=(12,7),color=["#342D7E","#008000"])
    pyplot.title("Model Losses for {}".format(CovModelNames[i]),fontsize=20,color=tlcol)
    pyplot.xlabel("Epoch",fontsize=20,color=tlcol)
    pyplot.ylabel("Loss",fontsize=20,color=tlcol)
    pyplot.grid()
    pyplot.show()
    print("||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||")
  
```

```python
CovResDF=pandas.DataFrame({"Model":CovModelNames,"Accuracy(Test)":CovACCts,"Loss(Test)":CovLossts,"Accuracy(Train)":CovACCtr,"Loss(Train)":CovLosstr})

for i in CovResDF.columns.tolist()[1:]:
    accout=CovResDF.sort_values(by=i,ascending=False)
    exvz = express.bar(accout, y=i, x="Model",color="Model",text=i,title="Comparison of {}".format(i),width=700)
    exvz.update_layout(
        font=dict(
            family="Times New Roman, Bold",
            size=20,
            color="black"
        )
    )
    exvz.show()
CovResDF=CovResDF.sort_values(by="Accuracy(Train)",ascending=False)
CovResDF
```
