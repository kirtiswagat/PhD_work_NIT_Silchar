```python
def expand_grayscale_image_channels(gray_image):
    try:
        gray_image_arr = np.array(gray_image)
        gray_image_arr = np.expand_dims(gray_image_arr, -1)
        gray_image_arr_3_channel = gray_image_arr.repeat(3, axis=-1)
        return gray_image_arr_3_channel
    except Exception as err:
        print('Info:', err)
```

```python
columns = ['Atelectasis', 'Cardiomegaly', 'Consolidation', 'Edema',
       'Pleural_Effusion', 'Pleural_Other', 'Pneumonia', 'Pneumothorax',
       'Enlarged_Cardiomediastinum', 'Lung_Opacity', 'Lung_Lesion', 'Fracture',
       'Support_Devices', 'No_Finding']
```

```python
datagen = ImageDataGenerator(rotation_range=15,
                            brightness_range= [0.6,1],
                            width_shift_range = 0.1,
                            height_shift_range = 0.1,
                            horizontal_flip = False,
                            vertical_flip= False,
                            preprocessing_function= preprocess_input)
```

```python
test_datagen = ImageDataGenerator(preprocessing_function= preprocess_input)
```

```python
train_generator = datagen.flow_from_dataframe(dataframe=train_df[:50000],directory="/kaggle/input/chexpert", x_col="Path",y_col=columns,batch_size=64,seed=42,shuffle=True,
                                                class_mode="raw",target_size=(150,150))
```

```python
valid_generator = datagen.flow_from_dataframe(dataframe=validation_df[:15000],directory="/kaggle/input/chexpert", x_col="Path",y_col=columns,batch_size=64,seed=42,shuffle=True,
                                                class_mode="raw",target_size=(150,150))
```

```python
test_generator = test_datagen.flow_from_dataframe(dataframe=test_df,directory="/kaggle/input/chexpert", x_col="Path",y_col=columns,batch_size=64,seed=42,shuffle=True,
                                                class_mode="raw",target_size=(150,150))
```

```python
conv_base = VGG16(include_top = False, weights='/kaggle/input/keras-pretrained-models/vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5', input_shape = (150,150,3))
```

```python
for layer in conv_base.layers[:-4]:
    layer.trainable = False
```

```python
top_model = conv_base.output
top_model = Flatten(name="Flatten")(top_model)
top_model = Dense(1024, activation = 'relu', name= 'Dense_1024')(top_model)
top_model = BatchNormalization()(top_model)
top_model = Dense(512, activation='relu', name= 'Dense_512')(top_model)
output_layer = Dense(14, activation = 'softmax',name= 'Output_Layer')(top_model)
```

```python
model = Model(inputs= conv_base.input, outputs = output_layer)
```

```python
model.summary()
```

```python
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size # 781
STEP_SIZE_VALID=valid_generator.n//valid_generator.batch_size #234
STEP_SIZE_TEST=test_generator.n//test_generator.batch_size #3
```

```python
optim_1 = Adam(learning_rate = 0.0001)
```

```python
model.compile(optimizer = optim_1, loss = "binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit_generator(generator=train_generator,steps_per_epoch=STEP_SIZE_TRAIN, validation_data=valid_generator, validation_steps=STEP_SIZE_VALID, epochs=50)
```
